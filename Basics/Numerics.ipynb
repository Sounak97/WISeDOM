Type Conversion : 

1) Implicit - In Implicit type conversion, Python automatically converts one data type to another. This process doesn't need any user involvement.Python avoids the loss of data in Implicit Type Conversion.

2) Explicit - (type casting) User intervention is used to convert one data type tp another. Enforcing the object to a specific datatype may end up with the loss of data.

```python
#Implicit - Example 1
c= a+b
print("Value: ", c)
print ("Type: ",type(c))
print(type(a))
print(type(b))

#Implicit Example 2
'''d = "20" #string data type
d = a+d
print(d)
#Will provide TypeError, for which we move on to explicit type conversion'''
```
